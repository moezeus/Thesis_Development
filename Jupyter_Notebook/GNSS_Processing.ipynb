```python
from IPython.display import Image
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
df = pd.read_csv("mengular.csv")
# df2 = pd.read_csv("samsung.csv")

plot_lim = [[np.min(df['long'])-0.00010, np.max(df['long']+0.00010)],
            [np.min(df['lat'])-0.00010, np.max(df['lat'])+0.00010]]

filename = 'movement_plot_animation'
buf_folder = filename+'_steps'
!mkdir $buf_folder
figsize=(9,9)
dpi=100
framecounter=0
redmi_long = []
redmi_lat = []
samsung_long = []
samsung_lat = []

print("Frame rendering started...")
for i in range(len(df['Time Stamp'])):
  # i = len(log["x_head"])-2
  fig = plt.figure(figsize=figsize, dpi=dpi)
  ax1 = fig.add_subplot(111)

  # Path visual
  redmi_long.append(df['long'][i])
  redmi_lat.append(df['lat'][i])

  # redmi_long.append(df['long'][i+50])
  # redmi_lat.append(df['lat'][i+50])

  # samsung_long.append(df2['long'][i+40])
  # samsung_lat.append(df2['lat'][i+40])

  ax1.scatter(x=redmi_long, y=redmi_lat, c='blue',marker='x', label="planting_gun")
  # ax1.scatter(x=samsung_long, y=samsung_lat, c='green',marker='s', label="planting_gun_2")
  
  # plt.grid(zorder=1)
  plt.xlabel("longitude")
  plt.ylabel("latitude")
  plt.xlim(plot_lim[0])
  plt.ylim(plot_lim[1])
  plt.legend()
  framecounter = framecounter + 1
  framename=buf_folder+'/'+'frame'+str(framecounter).zfill(10)+'.png'
  plt.savefig(framename)
  plt.close(fig)
print("Frame rendering finished.")
```
